#Exercise 1

```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
salaries
```

```python
#MEAN
n = len(salaries)

get_sum = sum(salaries)
mean = get_sum / n

print("Mean / Average is: " + str(mean))
```

```python
#Median
salaries.sort()
median_form = salaries[49] + salaries[50]
med = median_form / 2

print("The Median is: " + str(med))
```

```python
#Mode
salaries.sort()
list_1=[] # Create an empty list to store the frequency of each salary

# Iterate through the salary list
i=0
while i<len(salaries):
  list_1.append(salaries.count(salaries[i]))  # Count the occurrences of the current salary in the list and append to list
  i+=1 # increment the counter

dic = dict(zip(salaries, list_1)) # Create a dictionary to map salaries to their frequencies
dic2 = {k for (k,v)in dic.items() if v== max(list_1)} # Create a set to store salaries with the maximum frequency

print("The mode for this dataset: " + str(dic2))
```

```python
#Sample Variance
summation = 0
for x in salaries:
  summation += (x - mean)**2

sample_var = summation/(len(salaries)-1)

print("Sample Variance for this dataset: " + str(sample_var))
```

```python
#Sample Standard Deviation
stdev = sample_var**0.5
print("Standard Deviation for this dataset: " + str(stdev))
```
